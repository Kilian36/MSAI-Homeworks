```python
import numpy as np
import scipy
import matplotlib
from matplotlib import pyplot as plt


#Hola cabron
def linear_function(matrix, solution_array):
    #I compute the dot product between the matrix and the solution x in order
    #to obtain the vector b
    b=matrix.dot(solution_array)

    #I compute the condition number using np.linalg.cond
    cond_2norm=np.linalg.cond(matrix)
    cond_inf_norm=np.linalg.cond(matrix, np.inf)
    
    #I solve the linear system using np.linalg.solve
    x_computed=np.linalg.solve(matrix,b)

    #I compute the relative error between the two solution 
    #the function norm computes the 2-norm automatically for vectors
    rel_error=np.linalg.norm(x_computed-solution_array) / np.linalg.norm(solution_array)
    return cond_2norm, cond_inf_norm, rel_error
def plot(dimension, condInf, cond2, rel_error):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.plot(dimension,cond2)
    plt.plot(dimension, condInf)
    plt.grid()
    plt.title('Condition Number')
    plt.legend(['2-Norm', 'Inf-Norm'])
    plt.subplot(1,2,2)
    plt.plot(dimension, rel_error)
    plt.title('Relative error')
    plt.grid()
```

```python

dim=np.arange(10,110,10,int)
cond_2_list=[]
cond_inf_list=[]
rel_errors=[]
print(dim)
for el in dim:
    matrix=np.random.rand(el,el)
    x_true=np.ones(el)
    cond_2norm, cond_inf_norm, rel_error= linear_function(matrix, x_true)
    cond_2_list.append(cond_2norm)
    cond_inf_list.append(cond_inf_norm)
    rel_errors.append(rel_error)
plot(dim, cond_inf_list, cond_2_list, rel_errors)
```

```python
dim=np.arange(5,35,5,int)
cond_2_list=[]
cond_inf_list=[]
rel_errors=[]
for el in dim:
    x_true=np.ones(el)
    x=np.arange(1,el+1,1)
    matrix=np.vander(x)
    cond_2norm, cond_inf_norm, rel_error= linear_function(matrix, x_true)
    cond_2_list.append(cond_2norm)
    cond_inf_list.append(cond_inf_norm)
    rel_errors.append(rel_error)
plot(dim, cond_inf_list, cond_2_list, rel_errors)
```

```python
import scipy 
import scipy.linalg
dim=np.arange(4,13,1,int)
cond_2_list=[]
cond_inf_list=[]
rel_errors=[]
for el in dim:
    matrix=scipy.linalg.hilbert(el)
    x_true=np.ones(el)
    cond_2norm, cond_inf_norm, rel_error= linear_function(matrix, x_true)
    cond_2_list.append(cond_2norm)
    cond_inf_list.append(cond_inf_norm)
    rel_errors.append(rel_error)
plot(dim, cond_inf_list, cond_2_list, rel_errors)
```

```python
n_values = np.arange(1,100000,5000)
an = []
anBig = []
for n in n_values:
    an.append((1 + 1/n)**n)
exponents = np.arange(1,20,1)
num = 1000.0
new_values = []
for exp in exponents:
    value=num ** exp
    new_values.append(value)
new_n=np.array(new_values) 
for value in new_values:
    anBig.append((1 + 1/value)**value)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.xlabel("n")
plt.ylabel("An")
plt.grid()
plt.title("Nepero succession in function of n : Small n")
plt.plot(n_values, an)

plt.subplot(1,2,2)
plt.ylabel("An")
plt.xscale('log')
plt.title("Nepero succession in function of n : Big n")
plt.plot(new_values, anBig)
```
